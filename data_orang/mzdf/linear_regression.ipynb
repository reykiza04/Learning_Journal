# Library

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Seed Buat Konsistensi

```python
np.random.seed(0)
```

# Toy Dataset

```python
n_sample = 100
x = np.arange(0, n_sample) + np.random.normal(0, 8, n_sample)

plt.scatter(np.arange(n_sample), x);
```

# Model Definisi

```python
def forward(x, w, b):
    return x * w + b
```

# Parameter Definisi

```python
w = np.random.randn(1)
b = 0.

print(f"w: {w[0]}, b: {b}")
```

# Sebelum Training

```python
plt.plot(np.arange(n_sample), x, 'o')
plt.plot(np.arange(n_sample), np.arange(n_sample) * w + b)
```

# Training/Fitting

```python
def loss(x, w, b):
    return np.mean((forward(x, w, b) - x) ** 2)

def gradient_decent(x, w, b, learning_rate):
    w_grad = np.mean(2 * x * (forward(x, w, b) - x))
    b_grad = np.mean(2 * (forward(x, w, b) - x))
    w -= learning_rate * w_grad
    b -= learning_rate * b_grad
    return w, b

for i in range(100):
    w, b = gradient_decent(x, w, b, 0.0001)
print(f"w: {w[0]:.3f}, b: {b:.3f}, loss: {loss(x, w, b):.3f}")
```

# Setelah Training

```python
plt.plot(np.arange(n_sample), x, 'o')
plt.plot(np.arange(n_sample), np.arange(n_sample) * w + b)
```
