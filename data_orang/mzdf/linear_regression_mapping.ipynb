# Library

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Seed Buat Konsistensi

```python
np.random.seed(0)
```

# Toy Dataset

```python
n_sample = 100
x = np.arange(0, n_sample) + np.random.normal(0, 8, n_sample)
y = x * (x / 1)

plt.scatter(np.arange(n_sample), x)
plt.scatter(np.arange(n_sample), y);
```

# Model Definisi

```python
def forward(x, w, b):
    return x * w + b
```

# Parameter Definisi

```python
w = np.random.randn(1)
b = 0.

print(f"w: {w[0]}, b: {b}")
```

# Sebelum Training

```python
plt.plot(np.arange(n_sample), y, 'o', label='y')
plt.plot(np.arange(n_sample), forward(x, w, b), 'o', label='prediction')
plt.legend()
```

# Training/Fitting

```python
def loss(x, y, w, b):
    return np.mean((y - forward(x, w, b)) ** 2)


def gradient_decent(x, y, w, b, learning_rate):
    for i in range(1000):
        l = loss(x, y, w, b)
        w = w - learning_rate * np.mean(2 * x * (forward(x, w, b) - y))
        b = b - learning_rate * np.mean(2 * (forward(x, w, b) - y))
        if i % 100 == 0:
            print(f"loss: {l}, w: {w[0]}, b: {b}")
    return w, b, l


w, b, l = gradient_decent(x, y, w, b, 1e-4)
print(f"w: {w[0]:.3f}, b: {b:.3f}, loss: {l:.3f}")
```

# Setelah Training

```python
plt.plot(np.arange(n_sample), y, 'o', label='y')
plt.plot(np.arange(n_sample), forward(x, w, b), 'o', label='prediction')
plt.legend()
```
