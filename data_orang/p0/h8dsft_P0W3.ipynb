## Utility Function

```python
def plot_line(label, x, color, _ax):
    _ax.axvline(x, color=color, linestyle='--', linewidth=2, label=label)
    _ax.legend()
```

```python
def box_cox_transform(x, _l):
    if _l == 0:
        return np.log(x)
    return (x ** _l - 1) / _l


def inverse_box_cox_transform(x, _l):
    if _l == 0:
        return np.exp(x)
    return (x * _l + 1) ** (1 / _l)
```

> ## Data Preprocessing
> to better understand the data, we will perform some data preprocessing. in this case we will be using Power Transform to transform the data. to be more gausian, we will use the Box-Cox transform.

```python
l = .1
data_pledged = df[df.status == 'successful'].pledged

q1, q2, q3 = np.percentile(data_pledged, [25, 50, 75])
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) + q1
upper_bound = q3 + (1.5 * iqr) + q1
print(f'lower_bound: {lower_bound}')
print(f'upper_bound: {upper_bound}')

data_pledged_norm = box_cox_transform(data_pledged, l)
```

```python
def plot_distribution():
    _fig, _ax = plt.subplots(1, 2, figsize=(8 * 3, 8))
    _fig.suptitle('Pledged Amount')

    data = data_pledged[data_pledged < upper_bound]

    _ax[0].set_title('Distribution')
    sns.histplot(data, kde=True, bins=100, ax=_ax[0])

    _ax[1].set_title('Distribution (Box-Cox Transform)')
    sns.histplot(data_pledged_norm, kde=True, bins=100, ax=_ax[1])

    return _ax


plot_distribution();
```

## Measure of Central Tendency
> ### Mean
>the sum of all measurements divided by the number of observations in the data set.

```python
def calc_mean(x):
    return np.mean(x)


mean_pledged = calc_mean(data_pledged)
mean_pledged_norm = calc_mean(data_pledged_norm)

print(f"Mean: {mean_pledged}")
print(f"Mean (Box-Cox Transform): {mean_pledged_norm}")

ax = plot_distribution()

plot_line('Mean %.2f' % mean_pledged, mean_pledged, 'red', _ax=ax[0])
plot_line('Mean %.2f (%.2f)' % (mean_pledged_norm, mean_pledged), mean_pledged_norm, 'red', _ax=ax[1])
```

>>> #### Median
>>> the middle value that separates the higher half from the lower half of the data set. The median and the mode are the only measures of central tendency that can be used for ordinal data, in which values are ranked relative to each other but are not measured absolutely.

```python
def calc_median(x):
    return np.median(x)


median_pledged = calc_median(data_pledged)
median_pledged_norm = calc_median(data_pledged_norm)

ax = plot_distribution()

plot_line('Median %.2f' % median_pledged, median_pledged, 'red', _ax=ax[0])
plot_line('Median %.2f (%.2f)' % (median_pledged_norm, median_pledged), median_pledged_norm, 'red', _ax=ax[1])
```

>>> #### Mode Visualization
>>> the most frequent value in the data set. This is the only central tendency measure that can be used with nominal data, which have purely qualitative category assignments.

```python
def calc_mode(x):
    vals, counts = np.unique(x, return_counts=True)
    return vals[np.argmax(counts)]


mode_pledged = calc_mode(data_pledged)
mode_pledged_norm = calc_mode(data_pledged_norm)

print(f'mode_pledged: {mode_pledged}')
print(f'mode_pledged_norm: {mode_pledged_norm}')

ax = plot_distribution()

plot_line('Mode %.2f' % mode_pledged, mode_pledged, 'red', _ax=ax[0])
plot_line('Mode %.2f (%.2f)' % (mode_pledged_norm, mode_pledged), mode_pledged_norm, 'red', _ax=ax[1])
```

> ## All Measures of Central Tendency

```python
def calc_central_tendency(x):
    return {
        'mean': calc_mean(x),
        'median': calc_median(x),
        'mode': calc_mode(x)
    }


result = calc_central_tendency(data_pledged)
result_norm = calc_central_tendency(data_pledged_norm)

ax = plot_distribution()

mean_pledged = result['mean']
mean_pledged_norm = result_norm['mean']

median_pledged = result['median']
median_pledged_norm = result_norm['median']

mode_pledged = result['mode']
mode_pledged_norm = result_norm['mode']

plot_line('Mean %.2f' % mean_pledged, mean_pledged, 'r', _ax=ax[0])
plot_line('Mean %.2f (%.2f)' % (mean_pledged_norm, mean_pledged), mean_pledged_norm, 'r', _ax=ax[1])

plot_line('Median %.2f' % median_pledged, median_pledged, 'g', _ax=ax[0])
plot_line('Median %.2f (%.2f)' % (median_pledged_norm, median_pledged), median_pledged_norm, 'g', _ax=ax[1])

plot_line('Mode %.2f' % mode_pledged, mode_pledged, 'b', _ax=ax[0])
plot_line('Mode %.2f (%.2f)' % (mode_pledged_norm, mode_pledged), mode_pledged_norm, 'b', _ax=ax[1])
```

## Measure of Variance
> ### Range
>In statistics, the range of a set of data is the difference between the largest and smallest values.[1] Difference here is specific, the range of a set of data is the result of subtracting the sample maximum and minimum.

```python
def calc_range(x):
    return np.max(x) - np.min(x)


print('Range: %.2f' % calc_range(data_pledged))
print('Range (Box-Cox Transform): %.2f' % calc_range(data_pledged_norm))
```

> ### Variance
>Unlike range and interquartile range, variance is a measure of dispersion that takes into account the spread of all data points in a data set. It’s the measure of dispersion the most often used, along with the standard deviation, which is simply the square root of the variance.

```python
def calc_dispersion(x):
    return np.var(x)


print('Variance: %.2f' % calc_dispersion(data_pledged))
print('Variance (Box-Cox Transform): %.2f' % calc_dispersion(data_pledged_norm))
```

> ### Standard Deviation
>In statistics, the standard deviation is a measure of the amount of variation or dispersion of a set of values. A low standard deviation indicates that the values tend to be close to the mean (also called the expected value) of the set, while a high standard deviation indicates that the values are spread out over a wider range.

```python
def calc_deviation(x):
    return np.std(x)


print('Standard Deviation: %.2f' % calc_deviation(data_pledged))
print('Standard Deviation (Box-Cox Transform): %.2f' % calc_deviation(data_pledged_norm))
```

> ## All Measures of Variance

```python
def calc_variation(x):
    return {
        'range': calc_range(x),
        'dispersion': calc_dispersion(x),
        'deviation': calc_deviation(x)
    }


print("Pledged", calc_variation(df.pledged))
print("Pledged (Box-Cox Transform)", calc_variation(data_pledged_norm))
```

# Hypothesis Testing
> ## Hypothesis
> - H0: μ1 = μ2
> -  H1: μ1 != μ2
>> ## Null Hypothesis
>> Null Hypothesis is that the US Pledged mean is equal to the value of the GB Pledged mean parameter.
>> ## Alternative Hypothesis
>> The alternative hypothesis is the US Pledged mean is not equal to the value of the GB Pledged mean parameter.
> ## Confidence Interval
> In frequentist statistics, a confidence interval (CI) is a range of estimates for an unknown parameter, defined as an interval with a lower bound and an upper bound (notwithstanding one-sided confidence intervals, which are bounded only on one side). The interval is computed at a designated confidence level. The confidence level represents the long-run frequency of confidence intervals that contain the true value of the parameter. In other words, 95% of confidence intervals computed at the 95% confidence level contain the parameter, and likewise for other confidence levels.

```python
data_success = df[df.status == 'successful']
data_us_pledged = data_success[data_success.country == 'US'].pledged
data_gb_pledged = data_success[data_success.country == 'GB'].pledged

data_us_pledged_norm = box_cox_transform(data_us_pledged, l)
data_gb_pledged_norm = box_cox_transform(data_gb_pledged, l)

data_us_pledged_norm_sample = data_us_pledged_norm.sample(512, random_state=0)
data_gb_pledged_norm_sample = data_gb_pledged_norm.sample(512, random_state=0)

confidence_interval = .99
critical_value = 1. - confidence_interval

confidence_lower, confidence_upper = stats.norm.interval(confidence_interval, loc=data_us_pledged_norm.mean(), scale=data_us_pledged_norm.std())
t_stat, p_value = stats.ttest_ind(data_gb_pledged_norm_sample, data_us_pledged_norm_sample)

print("T-statistic: %.3f" % t_stat)
print("P-value: %.3f" % p_value)

print("\nConclusion:")
if p_value < critical_value:
    print("\tReject the null hypothesis")
else:
    print("\tFail to reject the null hypothesis")
```

> ## Confidence Interval Visualization
> after the t-test, we can see that our Alternative hypothesis is true. because the p-value is less than the critical value. or in the visualisation we can see that the confidence interval is not contain the true value of the parameter.

```python
plt.figure(figsize=(8 * 3, 8))
plt.title('US vs GB Pledged (Box-Cox Transform)')

plt.hist(data_us_pledged_norm, density=True, histtype='step', bins=100, label='US')
plt.hist(data_gb_pledged_norm, density=True, histtype='step', bins=100, label='GB')

plt.axvline(x=confidence_lower, color='g', linestyle='--', label='95% Confidence Interval')
plt.axvline(x=confidence_upper, color='g', linestyle='--')

plt.axvline(x=data_us_pledged_norm.mean() + t_stat * data_us_pledged_norm.std(), color='k', linestyle='--', label='Alternative Hypothesis')
plt.axvline(x=data_us_pledged_norm.mean() - t_stat * data_us_pledged_norm.std(), color='k', linestyle='--')
plt.axvline(x=data_us_pledged_norm.mean(), color='r', linestyle='--', label='Median')

plt.legend();
```

># T-Test Result
>after the t-test, we can see that our Alternative hypothesis is true. because the p-value is less than the critical value. or in the visualisation we can see that the confidence interval is not contain the true value of the parameter. so we can reject the null hypothesis. and we can say that the US Pledged mean is not equal to the value of the GB Pledged mean parameter.
