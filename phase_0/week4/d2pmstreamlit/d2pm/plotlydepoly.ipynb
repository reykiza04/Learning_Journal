# Data Cleaning

```python
df.drop(['Invoice ID','Branch','Tax 5%','Date','Time','gross margin percentage','gross income'],
       axis=1,inplace=True)
```

```python
df.head()
```

> will get the dummy variable format for the data

```python
df_dummies = pd.get_dummies(df)
```

```python
df_dummies.head()
```

```python
df_dummies = df_dummies.round(0)
```

# Model Training

> Now it's time to do a train test split, and train our model!

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_dummies.drop('Rating',axis=1)
y = df_dummies['Rating']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression(max_iter=10000)
```

```python
logmodel.fit(X_train,y_train)
```

# Predictions and Evaluations

>  Now predict values for the testing dat

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
