# Luas area lingkaran

```python
from scipy.integrate import trapz
import numpy as np

R=30
x=np.linspace(-R,R,200)

upper = np.sqrt(R**2 - x**2)
lower = -upper
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize=(6,6))
ax.plot(x,upper)
ax.plot(x,lower)
```

```python
area = trapz(upper,x) - trapz(lower,x)

print(area)
```
