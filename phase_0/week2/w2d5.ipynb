# Eigendecomposition

```python
import numpy as np

eig_val,eig_vec = np.linalg.eig(Green)

print('Eigen Vectors shape :', eig_vec.shape)
print('Eigen Values shape :', eig_val.shape)
```

```python
influencer = np.array([[0, 1/2, 1/2, 0, 0, 0, 0], #dari A
                       [0, 0, 0, 0, 0, 1, 0], #B
                       [1/2, 0, 0, 0, 1/2, 0, 0], #C
                       [0, 0, 1, 0, 0, 0, 0], #D
                       [1/3, 0, 1/3, 0, 0, 0, 1/3], #E
                       [0, 0, 0, 0, 1, 0, 0], #F
                       [0, 0, 0, 0, 0, 0, 0]]) #G

eigVal_inf,eigVec_inf = np.linalg.eig(influencer)
```

Output: menghitung pageRank

- cari eigen vector yang nilai eig val nya paling besar atau = 1
- cari nilai eigen vector yang paling besar, yang paling besar itu lah yang page rank nya bernilai paling tinggi dan merupakan influencer atau paling penting

```python
for i,val in enumerate(eigVal_inf):
  print('eigVal ke',i,':',val.real)
```

eigen vector ke 1 adalah eigen vector yang memiliki nilai eigen terbesar, vektor inilah yang mengandung nilai page rank masing-masing orang/agen

```python
eigVec_inf
```

```python
abc = ['A','B','C','D','E','F','G']
for i in range(7):
  print('Page Rank',abc[i],':',eigVec_inf[1,i].real)
```

Page Rank terbesar dimiliki oleh C dan D sehingga C dan D merupakan influencer di kelompok tersebut.

## Menghitung total pendapatan

Menghitung Total Pendapatan tahun 2019 di Supermarket Cabang Yangon

```python
import pandas as pd
import numpy as np
from scipy.integrate import trapz

data = pd.read_csv('https://raw.githubusercontent.com/fahmimnalfrzki/ds-masterclass-h8-vol2/main/data/supermarket_sales%20-%20Sheet1.csv')
data = data[data['City']=='Yangon'].sort_values(['Date','Time']).reset_index(drop=True)
data.head()
```

```python
data['gross income'].plot(kind='line')
# kita bisa menghitung total income menggunakan trapzoid 
```

```python
x = data.index
y = data['gross income']

total_gross_income_2019 = trapz(y,x)
print('Total Gross Income in 2019 (Integral):',total_gross_income_2019) # versi trapzoid dg cara menghitung luas
print('Total Gross Income in 2019 (pd.Series.sum()):',data['gross income'].sum()) #versi pandas dg car di tambah
```

## Menghitung luas negara Latvia

menghitung data yang ada garis lintangnya long (garis bujur) lat(garis lintang)

```python
import pandas as pd
import numpy as np

data_geo_map = pd.read_csv('https://public.opendatasoft.com/explore/dataset/world-administrative-boundaries/download/?format=csv&timezone=Asia/Jakarta&lang=en&use_labels_for_header=true&csv_separator=%3B',delimiter=';')
data_geo_map.head()
```

```python
Latvia = data_geo_map[data_geo_map['English Name']=='Latvia']['Geo Shape'].iloc[0]
Latvia

# dapat lat long di plot menjadi bentuk negara latvia, nah  kita akan menghitung luas negara latvia
```

```python
from ast import literal_eval

latvia_coor_dict = literal_eval(Latvia)

latvia_coor = np.array(latvia_coor_dict['coordinates'])
```

```python
latvia_coor = latvia_coor.reshape(latvia_coor.shape[1],latvia_coor.shape[2]).T

latvia_coor = pd.DataFrame({'long':latvia_coor[0],'lat':latvia_coor[1]})
latvia_coor.head()
```

```python
import matplotlib.pyplot as plt

plt.plot(latvia_coor.long,latvia_coor.lat)
```

kan dia menghitung luas jadi datanya akan di bagi dua atas dan bawah
jadi akan di hitung luas ke bawah dan atas di kurangin luas yg bawah

```python
median = latvia_coor.lat.median() # ini adalah garis tengah, di bagi dong
median
```

```python
from scipy.integrate import trapz

latvia_coor_sorted = latvia_coor.sort_values('long',ascending=True)

northern = latvia_coor_sorted[latvia_coor_sorted.lat >= median] # ini bagian atas
southern = latvia_coor_sorted[latvia_coor_sorted.lat < median] # ini bagian bawah
# baru akan di trapz
area_N = trapz(northern.lat,northern.long) #trapz(Y,X)
area_S = trapz(southern.lat,southern.long)

print('Luas area Latvia:',np.abs(area_N - area_S)) # agar arahnya positif pake abs absolut alias pasti positiv
```
