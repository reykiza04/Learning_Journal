```python
import numpy as np
import matplotlib.pyplot as plt

image = plt.imread('anjing_original.jpg')
plt.imshow(image)
```

```python
image
```

```python
image.shape
```

```python
Red = image[:,:,0] # semua baris dan kolom chanel 1
Green = image[:,:,1] # semua baris dan kolom chanel 2
Blue = image[:,:,2] # semua baris dan kolom chanel 3

print('Shape R:',Red.shape)
print('Shape G:',Green.shape)
print('Shape B:',Blue.shape)
```

```python
fig, ax = plt.subplots(ncols=4,figsize=(21,5))
ax[0].imshow(image)
ax[0].set_title('Gambar Asli')
ax[1].imshow(Red)
ax[1].set_title('Gambar Filter R')
ax[2].imshow(Green)
ax[2].set_title('Gambar Filter G')
ax[3].imshow(Blue)
ax[3].set_title('Gambar Filter B')
```

```python
grey = 0.299*Red + 0.587*Green + 0.114*Blue

dx, dy = np.gradient(grey)
#mencari dxdy disini

fig, ax = plt.subplots(ncols=3,figsize=(16,5))

ax[0].imshow(grey, cmap='gray')
ax[0].set_title('Real Image')

ax[1].imshow(dx, cmap='gray')
ax[1].set_title(r'$\partial_x img$')

ax[2].imshow(dy, cmap='gray')
ax[2].set_title(r'$\partial_y img$')
```
